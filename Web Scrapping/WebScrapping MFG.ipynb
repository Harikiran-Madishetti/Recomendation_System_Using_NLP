```python
import requests
import urllib
from bs4 import BeautifulSoup
import string
import re
import pandas as pd
```

```python
def Manufacture_page(Manufacturer_url):
    Company_Name= []
    Phone_Number= []
    Company_Address = []
    Company_State = []
    company_Website = []
    Company_Description =[]
    capabilities_list =[]
    Employee_Count =[]
    
    
    with urllib.request.urlopen('https://www.mfg.com'+ Manufacturer_url) as response:
        html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    
    if(soup.find(class_= "main").find('h1')):
        Company_Name = soup.find(class_= "main").find('h1').get_text().strip()
        
    if(soup.find(class_= "modal-body")):
        Phone_Number = soup.find(class_= "modal-body").get_text().strip()
        Phone_Number =re.sub("Call","",Phone_Number)
        
    if (soup.find(class_="col-md-12 address")):
        Company_Address = soup.find(class_="col-md-12 address").get_text().strip()
        Company_Address =re.sub("\n\t\t\t\t\t\tMAP","",Company_Address)
        Company_State= Company_Address.split(", ")[-1].split(" ")[0]
        
    if (soup.find(class_="col-md-12 visit-website")):
        company_Website = soup.find(class_="col-md-12 visit-website").find("a").get('href')
    
    if(soup.find(class_="col-md-12 description")):
        Company_Description = soup.find(class_="col-md-12 description").get_text().strip()
        
    if(soup.find_all(class_="col-md-12 capability")):    
        Machining_Capabilities =soup.find_all(class_="col-md-12 capability")[:-1]
        
        for i in Machining_Capabilities:
            if re.findall("^Machining.*",i.find("h3").get_text()):
                string_temp = i.get_text().strip()
                string_temp=re.sub('\s+', '', string_temp)        
        capabilities_list.append(string_temp)
        
    if(soup.find_all(class_="col-md-3 company-property")):   
            temp_No_of_Employee =soup.find_all(class_="col-md-3 company-property")

            for i in temp_No_of_Employee:
                if re.findall("^# Employees.*",i.find("h4").get_text()):
                    string_temp = i.find(class_="details").get_text().strip()       
            Employee_Count.append(string_temp)
            
    return   Company_Name, Phone_Number, Company_Address, Company_State, company_Website, Company_Description, capabilities_list, Employee_Count
         

        
    
```

```python
def main():
    global Supplier
    Supplier = pd.DataFrame() 
#     manufacturer_url_Machining
    df = pd.read_csv("manufacturer_url_Machining.csv",header=None)[0]
    Links_of_Manufactureres = df.values.tolist()
    
    i=0
    for url in Links_of_Manufactureres:
        
        i+=1
        print(i)
        print(url)
        Manufacturer_details = Manufacture_page(url)
        supplier = {}
        supplier['Company_Name'] = Manufacturer_details[0]
        supplier['Phone_Number']= Manufacturer_details[1]
        supplier['Company_Address'] = Manufacturer_details[2]
        supplier['Company_State'] = Manufacturer_details[3]
        supplier['company_Website'] = Manufacturer_details[4]
        supplier['Company_Description'] = Manufacturer_details[5]
        supplier['capabilities_list'] = Manufacturer_details[6]
        supplier['Employee_Count'] = Manufacturer_details[7]
 
    
        Supplier = Supplier.append(supplier, ignore_index = True)
        Supplier= Supplier[['Company_Name','Phone_Number','Company_Address', 'Company_State','company_Website','Company_Description','capabilities_list','Employee_Count']]

        Supplier.to_csv('MFG_Machining.csv')

    
    
```

```python
main()
```
